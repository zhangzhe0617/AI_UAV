```python
import pandas as pd
df = pd.read_excel("https://github.com/chris1610/pbpython/blob/master/data/sample-salesv3.xlsx?raw=True")
```

```python
df.head(10)
```

### 统计每人的总共ext price

```python
#最简单的对df进行groupby操作，然后进行一层统计
group1=df.groupby('name')
group1['ext price'].sum()
#从实现上看，groupby返回的是一个DataFrameGroupBy结构，这个结构必须调用聚合函数（如sum）之后，才会得到结构为Series的数据结果。
```

### 统计'ext price'这个属性在每个月的累和(sum)值，resample 只有在index为date类型的时候才能用

```python
df["date"] = pd.to_datetime(df['date'])#转换成datatime格式，便于后续再重采样
df.head(10)
```

```python
s=df.set_index('date')['ext price']
s.resample('M').sum()
```

```python
#等效于以下操作
serise=df['ext price']  
serise.index=df['date']
a=serise.resample('M').sum()
```

### 进一步的，我们想知道每个用户每个月的sum值，那么就需要一个groupby了：



```python
group=df.set_index('date').groupby('name')
group['ext price'].resample('M').sum()
```

```python
#另一种更简洁的写法
df.groupby(['name', pd.Grouper(key='date', freq='M')])['ext price'].sum()
#这种写法最简单，因为我们首先想到的是按照姓名groupby，然后再解决如何按照月统计总数
```

```python
df.groupby(['name', pd.Grouper(key='date', freq='A-DEC')])['ext price'].sum()
# 按照年度，且截止到12月最后一天统计ext price的sum值
```

## agg
### agg它提供基于列的聚合操作。而groupby可以看做是基于行，或者说index的聚合操作。

```python
df[["ext price", "quantity", "unit price"]].agg(['sum', 'mean'])
```

```python
# 等效于以下写法，但是更加简单便捷
df[['ext price','quantity','unit price']].sum()
```

```python
#可以针对不同的列使用不同的聚合函数：
df.agg({'ext price': ['sum', 'mean'], 'quantity': ['max', 'min'], 'unit price': ['mean']})
```
